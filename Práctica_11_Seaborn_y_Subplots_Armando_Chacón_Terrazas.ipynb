# **Importar librerías**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Cargamos el dataset de penguins**

```python
# Cargamos el dataset de penguins que incluye información sobre diferentes especies
df = sns.load_dataset("penguins")
```

# **Usa plt.subplots() para crear una figura con 2 filas y 2 columnas y gráficos**

```python
# Creamos una figura con 4 subplots organizados en 2 filas y 2 columnas
fig, ax = plt.subplots(2, 2, figsize=(15, 12))

# Primer gráfico (0,0): Histograma de las longitudes de aletas
sns.histplot(data=df, x='flipper_length_mm', ax=ax[0, 0], color='skyblue', kde=True)
ax[0, 0].set_title('Histograma de las longitudes de aletas', fontsize=14, fontweight='bold')
ax[0, 0].set_xlabel('Longitud de Aleta (mm)')
ax[0, 0].set_ylabel('Frecuencia')

# Segundo gráfico (0,1): Boxplot comparando la masa corporal por especie
# Añadimos hue='species' y legend=False para evitar el warning
sns.boxplot(data=df, x='species', y='body_mass_g', hue='species',
            ax=ax[0, 1], palette=['lightcoral', 'lightgreen', 'lightblue'], legend=False)
ax[0, 1].set_title('Boxplot comparando la masa corporal por especie', fontsize=14, fontweight='bold')
ax[0, 1].set_xlabel('Especie')
ax[0, 1].set_ylabel('Masa Corporal (g)')
ax[0, 1].tick_params(axis='x', rotation=45)

# Tercer gráfico (1,0): Dispersión entre masa corporal y largo de aletas
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g',
                hue='species', ax=ax[1, 0], palette='viridis', s=60)
ax[1, 0].set_title('Dispersión entre masa corporal y largo de aletas', fontsize=14, fontweight='bold')
ax[1, 0].set_xlabel('Longitud de Aleta (mm)')
ax[1, 0].set_ylabel('Masa Corporal (g)')
ax[1, 0].legend(title='Especie')

# Cuarto gráfico (1,1): Violin plot de la masa corporal según la especie
# Añadimos hue='species' y legend=False para evitar el warning
sns.violinplot(data=df, x='species', y='body_mass_g', hue='species',
               ax=ax[1, 1], palette=['lightcoral', 'lightgreen', 'lightblue'], legend=False)
ax[1, 1].set_title('Violin plot de la masa corporal según la especie', fontsize=14, fontweight='bold')
ax[1, 1].set_xlabel('Especie')
ax[1, 1].set_ylabel('Masa Corporal (g)')
ax[1, 1].tick_params(axis='x', rotation=45)

# Ajustar el diseño con plt.tight_layout()
plt.tight_layout()

# Añadimos un título general a toda la figura
plt.suptitle('Práctica #11: Seaborn y Subplots', fontsize=16, fontweight='bold', y=1.02)

# Mostramos la figura con todos los gráficos
plt.show()
```
