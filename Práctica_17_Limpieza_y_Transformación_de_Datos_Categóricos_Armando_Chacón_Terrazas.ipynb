```python
# Montar Google Drive para acceder a los archivos
from google.colab import drive
drive.mount('/content/drive')

# Importar pandas para trabajar con datos
import pandas as pd

# Cargar el archivo CSV desde Google Drive
# El archivo se encuentra en la ruta especificada
df = pd.read_csv('/content/drive/MyDrive/Unidad02/nls97.csv')

# Mostrar las primeras filas del dataset para verificar la carga
df.head()
```

# Ejercicio 01: Errores Tipográficos y Valores Inconsistentes: maritalstatus

```python
# 1. Mostrar valores únicos de la columna maritalstatus, correr el archivo solo una vez por que cambia lo que se imprime.
print("Valores únicos originales:")
print(df['maritalstatus'].unique())

# 2. Mostrar frecuencia de cada valor único
print("\nFrecuencia de valores originales:")
print(df['maritalstatus'].value_counts())

# 3. Cambiar el valor 'Never-married' a 'Single' (usando el valor EXACTO que aparece en los datos, en este caso cuidar la M con m)
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')

# 4. Mostrar frecuencia después deL cambio
print("\nFrecuencia después del cambio:")
print(df['maritalstatus'].value_counts())
```

# **Ejercicio 02: Errores Tipográficos y Valores Inconsistentes: highestdegree**

```python
# 1. Mostrar valores únicos originales
print("Valores únicos originales:")
print(df['highestdegree'].unique())

# 2. Mostrar frecuencia original
print("\nFrecuencia original:")
print(df['highestdegree'].value_counts())

# 3. Limpiar y normalizar los valores
df['highestdegree'] = (
    df['highestdegree']
    .str.lower()  # Convertir a minúsculas
    .str.replace(r'^\d+\.\s*', '', regex=True)  # Eliminar números y puntos al inicio
    .str.strip()  # Eliminar espacios sobrantes
)

# 4. Mostrar valores únicos después de la limpieza
print("\nValores únicos después de la limpieza:")
print(df['highestdegree'].unique())
```

# **Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas**

```python
# 1. Unir categorías 'Widowed' y 'Single' en nueva categoría
df['marital_grouped'] = df['maritalstatus'].replace({
    'Widowed': 'Single/No Partner',
    'Single': 'Single/No Partner'
})

# 2. Mover la nueva columna al principio del DataFrame
cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index('marital_grouped')))
df = df[cols]

# 3. Mostrar valores únicos de la nueva columna
print("Valores únicos de marital_grouped:")
print(df['marital_grouped'].unique())

# 4. Mostrar frecuencia de la nueva columna
print("\nFrecuencia de marital_grouped:")
print(df['marital_grouped'].value_counts())
```

# **Ejercicio 04: Tipos de Datos**

```python
# 1. Imprimir el tipo de datos de la columna 'gender'
print("Tipo de datos original de gender:")
print(df['gender'].dtype)

# 2. Imprimir la cantidad de bytes que ocupa la columna 'gender'
print("\nCantidad de bytes original de gender:")
print(df['gender'].memory_usage(deep=True))

# 3. Verificar si la columna es de tipo 'object' y cambiarla a 'category'
if df['gender'].dtype == 'object':
    df['gender'] = df['gender'].astype('category')
    print("\n✅ Columna gender convertida a category")

# 4. Imprimir la cantidad de bytes después del cambio (si se realizó)
print("\nCantidad de bytes después del cambio:")
print(df['gender'].memory_usage(deep=True))
```
