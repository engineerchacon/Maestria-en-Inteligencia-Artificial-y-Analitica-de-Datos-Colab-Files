# Ejercicio 01:Tipo de Datos Correcto

Carga el conjunto de datos airdata.csv y verifica que cada columna tenga el tipo de datos correcto
Si una columna de tipo temporal no está en el tipo de correcto realiza la conversión.
Usa dtypes para mostrar que cada columna, principalmente la que contiene datos temporales, está en el tipo de datos correcto
Imprime las primeras tres líneas


```python
# Importamos la biblioteca pandas, que nos ayuda a manipular datos tabulares
import pandas as pd

# Cargamos el archivo CSV desde la ruta en Drive
df = pd.read_csv('/content/drive/MyDrive/Unidad02/Airdata.csv')

# Mostramos los tipos de datos actuales de cada columna
print("Tipos de datos originales:")
print(df.dtypes)
print("\n")

# La columna 'DateTime' contiene fechas y horas, pero es de tipo 'object' (texto)
# Se convierte a un tipo de dato datetime para poder trabajar con ella correctamente
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Verificamos nuevamente los tipos de datos después de la conversión
print("Tipos de datos después de la conversión:")
print(df.dtypes)
print("\n")

# Mostramos las primeras 3 líneas del DataFrame para verificar que todo se vea bien
print("Primeras 3 líneas del conjunto de datos:")
print(df.head(3))
```

# Ejercicio 02: Extraer componentes de fecha y tiempo

De la columna 'DateTime' extrae el año, mes, día y hora. Crea una nueva columna para cada componente que extraes
Imprime las primeras tres líneas del dataframe que ahora deberá tener 4 nuevas columnas.


```python
# Extraemos componentes de fecha y hora de la columna DateTime
# y los agregamos al principio del DataFrame para mejor visualización

# Año: extraemos el año de cada fecha y lo insertamos en la posición 0 (primera columna)
df.insert(0, 'Year', df['DateTime'].dt.year)

# Mes: extraemos el mes de cada fecha y lo insertamos en la posición 1
df.insert(1, 'Month', df['DateTime'].dt.month)

# Día: extraemos el día del mes de cada fecha y lo insertamos en la posición 2
df.insert(2, 'Day', df['DateTime'].dt.day)

# Hora: extraemos la hora de cada fecha y lo insertamos en la posición 3
df.insert(3, 'Hour', df['DateTime'].dt.hour)

# Mostramos las primeras 3 filas del DataFrame con las nuevas columnas al principio
print("DataFrame con las nuevas columnas de componentes de fecha y hora:")
print(df.head(3))
```

# Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)

Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'. Guarda el resultado en una nueva columna y agrega al dataframe.
Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.


```python
# Sumamos 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'
# Usamos pd.Timedelta para crear el intervalo de tiempo especificado
df['DateTime_suma'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)

# Mostramos solo las columnas 'DateTime' y la nueva columna con la suma
print("Columna DateTime original y columna con la suma de 10 días, 7 horas y 15 minutos:")
print(df[['DateTime', 'DateTime_suma']].head())
```

# Ejercicio 04: Timestamp

Convierte la columna 'DateTime' la cual tiene un formato de fecha legible a a timestamp Unix. Guarda el resultado en una nueva columna
Imprime las primeras tres filas de  la columna 'DateTime' y la columna que representa dicha fecha en timestamp Unix


```python
# Convertimos la columna 'DateTime' a timestamp Unix usando el método dt.timestamp()
# Este método convierte directamente a segundos desde la época Unix (1970-01-01)
df['Unix_Timestamp'] = df['DateTime'].astype('int64') // 10**9

# Mostramos las primeras 3 filas de las columnas 'DateTime' y 'Unix_Timestamp'
print("Columna DateTime y su correspondiente timestamp Unix:")
print(df[['DateTime', 'Unix_Timestamp']].head(3))
```

# Ejercicio 05: Filtrado de Fechas

Utiliza la columna 'DateTime' y Filtra aquellas fechas que son mayores a 2020-11-01
Imprime el resultado
Sube un archivo txt que contenga la liga a tu notebook

```python
# Filtramos las fechas mayores a 2020-11-01
fecha_limite = pd.to_datetime('2020-11-01')
df_filtrado = df[df['DateTime'] > fecha_limite]

# Configuramos pandas para mostrar todas las filas
pd.set_option('display.max_rows', None)

# Mostramos el DataFrame filtrado
print("Filas con fechas mayores a 2020-11-01:")
print(df_filtrado)
```
