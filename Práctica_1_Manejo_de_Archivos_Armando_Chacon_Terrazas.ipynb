# **Crear una lista vacia para poder guardar los números**

```python
lista_1 = []
```

# **Abrir el archivo y buscar líneas**

```python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')

for linea in archivo:
    if linea.startswith('X-DSPAM-Confidence'):
        # Encontrar los dos puntos
        posicion = linea.find(':')
        # Tomar el número después de los dos puntos
        numero = linea[posicion+1:].strip()
        # Convertir a número decimal y guardar en lista_1
        lista_1.append(float(numero))

archivo.close()
```

# **Mostrar los números encontrados**

```python
print("Los números encontrados son:")
for num in lista_1:
    print(num)
```

# **Hacer los cálculos e imprimir**

```python
if len(lista_1) > 0:
    # 1. CONTAR ELEMENTOS
    cantidad = len(lista_1)

    # 2. SUMAR TODOS LOS VALORES
    suma = sum(lista_1)

    # 3. CALCULAR PROMEDIO
    promedio = suma / cantidad

    # 4. CALCULAR VARIANZA
    suma_cuadrados = 0  # Iniciamos un contador en cero

    for num in lista_1:
        diferencia = num - promedio  # Diferencia entre cada número y el promedio
        cuadrado = diferencia ** 2  # Elevamos esa diferencia al cuadrado
        suma_cuadrados += cuadrado  # Sumamos cada resultado al contador

    varianza = suma_cuadrados / cantidad  # Promedio de las diferencias al cuadrado


    # Imprimir
    print("\n--- RESULTADOS ---")
    print(f"Cantidad de datos extraídos: {cantidad}")
    print(f"Sumatoria de los datos: {round(suma, 4)}") # Redondeado a 4 decimales
    print(f"Promedio de los datos: {round(promedio, 4)}") # Redondeado a 4 decimales
    print(f"Varianza de los datos: {round(varianza, 4)}") # Redondeado a 4 decimales

else:
    print("\nNo se encontraron números X-DSPAM-Confidence")
```
