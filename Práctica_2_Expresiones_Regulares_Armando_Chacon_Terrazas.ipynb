# **Buscar el patrón exacto "Valor actual: $XXXX.XX"**

```python
print("-INICIANDO BÚSQUEDA-")  # Mostramos un mensaje de inicio

# Vamos a leer el archivo línea por línea, se tuvo que agregar la parte de encoding por que generaba error al momento de leer el archivo.
with open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt', encoding='latin-1') as archivo:
    for linea in archivo:

        #Quitamos espacios y saltos de línea al final
        linea_limpia = linea.rstrip()

        # Búsqueda del patrón "Valor actual: $XXXX.XX" en la línea actual
        # 're.search()' busca el patrón y'[0-9.]+' busca números del 0 al 9 ([0-9]), puntos (\.), que aparezcan 1 o más veces (+)
        if re.search(r'Valor actual: \$[0-9.]+', linea_limpia):

            # 're.findall()' encuentra todos los valores que coincidan con el patrón
            # Los paréntesis () en '([0-9.]+)' dicen: "solo guarda esta parte"
            valor = re.findall(r'Valor actual: \$([0-9.]+)', linea_limpia)
            # 'valor' es una lista como ['1338.97'], por eso usamos valor[0] para acceder al primer elemento

            # Mostramos el resultado encontrado
            print(f"${valor[0]}")

print("-BÚSQUEDA TERMINADA-")
```
