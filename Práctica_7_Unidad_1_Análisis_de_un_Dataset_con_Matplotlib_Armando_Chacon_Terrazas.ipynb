## **Importar librerias**

```python
#Importamos matplotlib.pyplot y con plt acortamos su nombre
import matplotlib.pyplot as plt

#Importamos seaborn y con sns acortamos su nombre
import seaborn as sns
```

# **Carga y exploración del dataset**

```python
#Cargar el dataset Iris utilizando seaborn o cualquier otro dataset de su elección
flores = sns.load_dataset('iris')
#Verificar las primeras filas del dataset con el método .head()
print("Ver primeras filas del dataset")
print(flores.head())  # Muestra las primeras 5 flores con todas sus medidas
#Obtener información general sobre los datos (nombres de columnas, tipos de datos, valores nulos, etc.)
#Listamos los nombres de todas las columnas (variables) disponibles
print("\nInformación general del dataset")
print("Nombres de columnas:", list(flores.columns))
#Mostrar tipos de datos
print("\nTipos de datos")
print(flores.dtypes)
#Mostrar valores nulos
print("\nValores nulos por columna:")
print(flores.isnull().sum())  # Suma de valores nulos por columna
```

# **Mismo tamaño para todos los gráficos (10 pulgadas de ancho x 6 de alto)**

```python
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.autolayout'] = True #Activar ajuste automático
```

# **a) Gráfico de barras:**

```python
#Creamos una nueva figura
plt.figure()

#Añadimos el título superior a) Gráfico de barras
plt.suptitle('a) Gráfico de barras', fontsize=14, fontweight='bold')

#Calcular los promedios
#Se agrupan las flores por especie y calculamos el promedio de:
# petal_length (largo del pétalo)
# petal_width (ancho del pétalo)
promedios = flores.groupby('species')[['petal_length', 'petal_width']].mean()

#Crear las barras para el largo del pétalo
#promedios.index: las especies (Setosa, Versicolor, Virginica) en el eje X
#promedios['petal_length']: los valores promedio en el eje Y
#color='gold': https://matplotlib.org/stable/gallery/color/named_colors.html
#label='Largo'
plt.bar(promedios.index, promedios['petal_length'], color='gold', label='Largo')

#Crear las barras para el ANCHO del pétalo
#Mismas especies en eje X
#Promedios['petal_width']: valores promedio de ancho
#color='salmon': barras color salmón
#alpha=0.7: 70% de opacidad para ver las barras detrás, https://matplotlib.org/stable/gallery/color/set_alpha.html
#label='Ancho': etiqueta para la leyenda
plt.bar(promedios.index, promedios['petal_width'], color='salmon', alpha=0.7, label='Ancho')

#Personalizar gráfico:
plt.title('Promedio de Medidas de Pétalos por Especie')
plt.xlabel('Especies')         # Etiqueta del eje horizontal (X)
plt.ylabel('Centímetros (cm)') # Etiqueta del eje vertical (Y)
plt.legend()                   # Muestra la leyenda (Largo/Ancho)

# Mostrar el gráfico de barras
plt.show()
```

# **b) Histograma:**

```python
#Creamos una nueva figura
plt.figure()

#Añadimos el título superior b) Histograma
plt.suptitle('b) Histograma', fontsize=14, fontweight='bold')

#Crear el histograma
#flores['sepal_length']: usamos los valores de largo de sépalo
#bins=15: dividimos los datos en 15 grupos (barras)
#color='lightgreen': color verde claro para las barras
#edgecolor='black': bordes negros para definir cada barra
plt.hist(flores['sepal_length'], bins=15, color='lightgreen', edgecolor='black')

#Personalizar gráfico:
plt.title('Distribución del Largo del Sépalo')  # Título principal
plt.xlabel('Largo (cm)')         # Eje X: valores de largo
plt.ylabel('Cantidad de Flores') # Eje Y: cuántas flores tienen cada medida
plt.grid(True, linestyle='--', alpha=0.3)  # Cuadrícula punteada suave

#Mostrar el histograma
plt.show()
```

# **c) Gráfico de dispersión (scatter plot) :**

```python
#Creamos una nueva figura
plt.figure()

#Añadimos el título superior c) Gráfico de dispersión
plt.suptitle('c) Gráfico de dispersión (scatter plot) ', fontsize=14, fontweight='bold')

#Preparar datos, obtenemos la lista de especies únicas (Setosa, Versicolor, Virginica)
especies = flores['species'].unique()

#Definimos colores diferentes para cada especie
colores = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

#Graficar cada especie por separado
for especie in especies:
    #Filtramos solo las flores de esta especie
    datos = flores[flores['species'] == especie]

    #Creamos los puntos de dispersión:
    #datos['petal_length']: valores para el eje X (largo del pétalo)
    #datos['petal_width']: valores para el eje Y (ancho del pétalo)
    #color: asignamos el color correspondiente a la especie
    #label: nombre de la especie para la leyenda
    #alpha=0.7: 70% de opacidad para ver puntos superpuestos
    plt.scatter(
        datos['petal_length'],
        datos['petal_width'],
        color=colores[especie],
        label=especie,
        alpha=0.7
    )

#Personalizar gráfico:
plt.title('Relación entre Largo y Ancho del Pétalo')  #Título principal
plt.xlabel('Largo del Pétalo (cm)')  #Eje X
plt.ylabel('Ancho del Pétalo (cm)')  #Eje Y
plt.legend()                         #Muestra leyenda con especies
plt.grid(True, linestyle='--', alpha=0.3)  #Cuadrícula suave

#Mostrar el gráfico de dispersión
plt.show()
```

# **d) Gráfico de cajas (box plot) :**

```python
#Creamos una nueva figura
plt.figure()

#Añadimos el título superior d) Gráfico de cajas(box plot)
plt.suptitle('d) Gráfico de cajas (box plot)', fontsize=14, fontweight='bold')

#Preparar datos
#Se crea una lista con tres grupos de datos (uno por especie)
#Cada grupo contiene todos los valores de sepal_length para esa especie
datos_grafico = [
    flores[flores['species'] == 'setosa']['sepal_length'],
    flores[flores['species'] == 'versicolor']['sepal_length'],
    flores[flores['species'] == 'virginica']['sepal_length']
]

#Crear gráfico de cajas
# datos_grafico: la lista con los tres grupos de datos
#patch_artist=True: permite colorear las cajas
#boxprops: propiedades de las cajas (color de relleno azul claro)
plt.boxplot(datos_grafico,
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            )

#Personalizar gráfico:
plt.title('Distribución del Largo del Sépalo por Especie')  # Título principal
plt.xlabel('Especies')          # Eje X: nombres de especies
plt.ylabel('Largo (cm)')        # Eje Y: valores de largo
plt.xticks([1, 2, 3], especies) # Asignar nombres a las marcas 1,2,3 del eje X
plt.grid(True, linestyle='--', alpha=0.3)  # Cuadrícula suave

#Mostrar el gráfico de cajas (box plot)
plt.show()
```

# **e) Gráfico de líneas**

```python
#Creamos una nueva figura
plt.figure()

#Añadimos el título superior e) Gráfico de líneas:
plt.suptitle('e) Gráfico de líneas', fontsize=14, fontweight='bold')

#Crear gráfico de líneas
#Graficamos todos los valores de petal_length en orden de aparición
#flores['petal_length']: valores en el eje Y
#color='purple': línea morada
#linewidth=1.5: grosor de la línea
plt.plot(flores['petal_length'], color='purple', linewidth=1.5)

#Personalizar gráfico:
plt.title('Variación del Largo del Pétalo')  # Título
plt.xlabel('Número de Flor (siguiendo el orden en el dataset)')  # Eje X: índice de la flor
plt.ylabel('Largo del Pétalo (cm)')          # Eje Y: valor del largo
plt.grid(True, linestyle='--', alpha=0.3)    # Cuadrícula suave

# Mostrar el gráfico de líneas
plt.show()
```
