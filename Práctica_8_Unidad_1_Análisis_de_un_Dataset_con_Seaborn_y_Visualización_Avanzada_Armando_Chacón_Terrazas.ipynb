# **Importar librerías**

```python
#Usamos 'sns' como nombre corto
import seaborn as sns
#Importamos matplotlib.pyplot para crear gráficos en Python
#Usamos 'plt' como nombre corto
import matplotlib.pyplot as plt
#Importamos numpy - Librería para operaciones matemáticas y cálculos numéricos
#Usamos 'np' como nombre corto
import numpy as np
```

# **Carga y exploración del dataset:**

```python
#Cargar el dataset tips usando seaborn.
propinas = sns.load_dataset('tips')

# Obtenemos información general sobre el dataset
print("\n-Visión general de los datos-")

#Realizar un análisis preliminar: utiliza .head(), .info(), .describe() para obtener una visión general de los datos.
# Imprimimos las primeras 5 filas del dataset con .head()
print("Información que recibimos utilizando .head()")
print(propinas.head())  # Muestra las primeras 5 transacciones

# Usamos .info() para obtener información detallada sobre la estructura del dataset
print("\nInformación que recibimos utilizando .info()")
propinas.info()

# Usamos .describe() para obtener estadísticas descriptivas de las columnas numéricas
print("\nInformación que recibimos utilizando .describe()")
print(propinas.describe())

# Verificamos si hay valores faltantes (nulos) en cada columna
print("\nValores nulos por columna:")
print(propinas.isnull().sum())  # Suma de valores nulos por columna
```

# **Mismo tamaño para todos los gráficos (12 pulgadas de ancho x 8 de alto)**

```python
#Escribimos el mismo tamaño para todos los gráficos (12 pulgadas de ancho x 8 de alto)
plt.rcParams['figure.figsize'] = (12, 8)
#Activamos el ajuste automático
plt.rcParams['figure.autolayout'] = True
#Configuramos el estilo de seaborn para todos los gráficos
sns.set_style("whitegrid")  # Fondo con cuadrícula blanca
```

# **a) Mapa de Calor (Heatmap):**

```python
# Creamos una nueva figura
plt.figure()

# Añadimos un título superior
plt.suptitle('a) Mapa de Calor - Correlación entre Variables Numéricas', fontsize=14, fontweight='bold')

#Calcular la matriz de correlación
#La correlación mide la relación entre variables -1 a 1, seleccionamos solo las columnas con numeros para calcular correlaciones
matriz_correlacion = propinas[['total_bill', 'tip', 'size']].corr()

#Crear el mapa de calor
#matriz_correlacion: los valores de correlación a visualizar
# annot=True: muestra los valores numéricos en cada celda
# cmap="coolwarm": paleta de colores (azul para negativas, rojo para positivas)
# center=0: centra la paleta en 0 (correlación neutra)
# square=True: hace que las celdas sean cuadradas
# linewidths=0.5: añade líneas delgadas entre celdas
sns.heatmap(matriz_correlacion,
            annot=True,
            cmap="coolwarm",
            center=0,
            square=True,
            linewidths=0.5)

# Personalizar gráfico
plt.title('Matriz de Correlación entre Variables Numéricas')  # Título principal

# Mostrar el gráfico en pantalla
plt.show()
```

# **b) Diagrama de Violín:**

```python
# Creamos una nueva figura
plt.figure()

# Título superior
plt.suptitle('b) Diagrama de Violín - Distribución de Propinas por Día', fontsize=14, fontweight='bold')

# Crear el diagrama de violín
# x="day": variable categórica en el eje X (días de la semana)
# y="tip": variable numérica en el eje Y (monto de la propina)
# data=propinas: el dataframe que contiene los datos
# hue="day": variable para colorear (misma que x para evitar warning)
# palette="pastel": paleta de colores suaves
# legend=False: oculta la leyenda redundante
sns.violinplot(x="day", y="tip", data=propinas, hue="day", palette="pastel", legend=False)

# Personalizar gráfico
plt.title('Distribución de Propinas por Día de la Semana')  # Título principal
plt.xlabel('Día de la Semana')  # Etiqueta del eje X
plt.ylabel('Monto de la Propina ($)')  # Etiqueta del eje Y

# Mostrar el gráfico
plt.show()
```

# **c) Gráfico de dispersión (Scatter Plot):**

```python
#Nueva figura
plt.figure()

#Título superior
plt.suptitle('c) Gráfico de Dispersión - Total de Cuenta vs Propina', fontsize=14, fontweight='bold')

#Crear el gráfico de dispersión
#Muestra la relación entre dos variables numéricas
# x="total_bill": variable en el eje X (total de la cuenta)
# y="tip": variable en el eje Y (monto de la propina)
# hue="day": colorea los puntos según el día de la semana
# data=propinas: el dataframe que contiene los datos
# palette="viridis": paleta de colores variada y atractiva
# alpha=0.7: transparencia para ver puntos superpuestos
sns.scatterplot(x="total_bill", y="tip", hue="day", data=propinas, palette="viridis", alpha=0.7)

# Personalizar gráfico
plt.title('Relación entre Total de Cuenta y Propina')  # Título principal
plt.xlabel('Total de la Cuenta ($)')  # Etiqueta del eje X
plt.ylabel('Monto de la Propina ($)')  # Etiqueta del eje Y
plt.legend(title='Día de la Semana')  # Leyenda con título

# Mostrar gráfico
plt.show()
```

# **d) Gráfico de barras (Bar Plot):**

```python
# Nueva figura
plt.figure()

# Título superior
plt.suptitle('d) Gráfico de Barras - Promedio de Propina por Género', fontsize=14, fontweight='bold')

#Crear el gráfico de barras
# x="sex": variable categórica en el eje X (género)
# y="tip": variable numérica en el eje Y (propina)
# data=propinas: el dataframe que contiene los datos
# estimator=np.mean: función para agregar los datos
# errorbar=None: no mostrar intervalos de confianza
# egend=False: oculta la leyenda
sns.barplot(x="sex", y="tip", data=propinas, hue="sex", palette=["lightblue", "pink"],
            estimator=np.mean, errorbar=None, legend=False)

# Personalizar gráfico
plt.title('Promedio de Propina por Género')  # Título
plt.xlabel('Género')  # Etiqueta del eje X
plt.ylabel('Propina Promedio ($)')  # Etiqueta del eje Y

# Mostrar gráfico
plt.show()
```
