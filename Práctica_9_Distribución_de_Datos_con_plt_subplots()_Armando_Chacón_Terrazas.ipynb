# **Importar librerías**

```python
# Importamos pandas para crear y manipular DataFrames
import pandas as pd
# Importamos numpy para generar distribuciones de datos
import numpy as np
# Importamos matplotlib.pyplot para crear visualizaciones
import matplotlib.pyplot as plt
```

# **Genera un DataFrame con Pandas**

```python
# Creamos un DataFrame con 4 columnas diferentes
# Cada columna tiene 1000 números
d_frame = pd.DataFrame({
    # COLUMNA "NORMAL":
    # np.random.normal(0, 1, 1000) genera 1000 números con distribución normal
    # - El primer número (0): es la "media" (el valor promedio alrededor del cual se agrupan los números)
    # - El segundo número (1): es la "desviación estándar" (qué tan dispersos están los números)
    # - El tercer número (1000): es la cantidad de números a generar
    'Normal': np.random.normal(0, 1, 1000),

    # COLUMNA "UNIFORME":
    # np.random.uniform(0, 1, 1000) genera 1000 números distribuidos uniformemente
    # - El primer número (0): es el valor mínimo que pueden tener los números
    # - El segundo número (1): es el valor máximo que pueden tener los números
    # - El tercer número (1000): es la cantidad de números a generar
    'Uniforme': np.random.uniform(0, 1, 1000),

    # COLUMNA "EXPONENCIAL":
    # np.random.exponential(1, 1000) genera 1000 números con distribución exponencial
    # - El primer número (1): es el parámetro de "escala" (afecta qué tan rápido decrece la curva)
    # - El segundo número (1000): es la cantidad de números a generar
    'Exponencial': np.random.exponential(1, 1000),

    # COLUMNA "POISSON":
    # np.random.poisson(5, 1000) genera 1000 números con distribución de Poisson
    # - El primer número (5): es el parámetro "lambda" significa que esperamos un promedio de 5 eventos por intervalo
    # - El segundo número (1000): es la cantidad de números a generar
    'Poisson': np.random.poisson(5, 1000)
})
```

# **Crea una figura con plt.subplots(2, 2)**

# **Grafica cada distribución en un subplot usando ax[i, j].hist().**

```python
# Creamos una figura con 4 gráficos organizados en 2 filas y 2 columnas
# fig es donde dibujaremos todos los gráficos
# ax es una "matriz" que nos permite acceder a cada gráfico individualmente
# figsize=(10, 8): define el tamaño (10 pulgadas de ancho, 8 de alto)
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# PRIMER GRÁFICO (posición [0, 0] - fila 0, columna 0)
# Creamos un histograma con los datos de la columna "Normal"
# - d_frame['Normal']: tomamos los datos de la columna "Normal"
# - bins=30: dividimos los datos en 30 grupos/barras
# - color='blue': las barras serán de color azul
# - alpha=0.7: las barras serán ligeramente transparentes (70% opacas)
ax[0, 0].hist(d_frame['Normal'], bins=30, color='blue', alpha=0.7)
# Añadimos un título al primer gráfico
ax[0, 0].set_title('Distribución Normal')

# SEGUNDO GRÁFICO (posición [0, 1] - fila 0, columna 1)
# Creamos un histograma con los datos de la columna "Uniforme"
ax[0, 1].hist(d_frame['Uniforme'], bins=30, color='green', alpha=0.7)
# Añadimos un título al segundo gráfico
ax[0, 1].set_title('Distribución Uniforme')

# TERCER GRÁFICO (posición [1, 0] - fila 1, columna 0)
# Creamos un histograma con los datos de la columna "Exponencial"
ax[1, 0].hist(d_frame['Exponencial'], bins=30, color='red', alpha=0.7)
# Añadimos un título al tercer gráfico
ax[1, 0].set_title('Distribución Exponencial')

# CUARTO GRÁFICO (posición [1, 1] - fila 1, columna 1)
# Creamos un histograma con los datos de la columna "Poisson"
ax[1, 1].hist(d_frame['Poisson'], bins=30, color='purple', alpha=0.7)
# Añadimos un título al cuarto gráfico
ax[1, 1].set_title('Distribución Poisson')

# Ajustamos automáticamente el espacio entre los gráficos
# Esto evita que los títulos, etiquetas o gráficos se superpongan
plt.tight_layout()

# Mostramos la figura con todos los gráficos en la pantalla
plt.show()
```
